```python
#Manage data for variable 2
```

```python
df['H4PE36'].value_counts()
```

```python
df['future'] = df['H4PE36']
df['future'].value_counts()
```

```python
df['future'] = df['future'].loc[df['future'] !='6']
df['future'] = df['future'].loc[df['future'] !='8']
df['future'] = df['future'].loc[df['future'] !=' ']
df['future'] = df['future'].loc[df['future'] !='.']
df['future'].value_counts()
```

```python
df['future'].replace(['4','5','3','2','1'] , ['Disagree','Strongly Disagree','Neither agree or disagree','Agree','Strongly agree'], inplace = True)
df['future'].value_counts()
```

```python
rp.summary_cat(df['future'])
```

```python
#Graphs and tables or statistics
```

```python
fig, ax = plt.subplots(figsize=(6, 6))

x = [61.36, 22.25, 9.29, 6.10, 1.00]
labels=["Disagree", "Strongly Disagree", "Neither Agree or Disagree", "Agree", "Strongly Agree"]

ax.pie(x, labels=labels, autopct='%.1f%%',
       wedgeprops={'linewidth': 3.0, 'edgecolor': 'white'},
       textprops={'size': 'x-large'})
ax.set_title('I live my life without much thought for the future', fontsize=18)
plt.tight_layout()
```

```python
df['future'].value_counts(normalize = True).plot(kind ='bar')
plt.title("I live my life without much thought for the future")
plt.ylabel("Percent")
```
